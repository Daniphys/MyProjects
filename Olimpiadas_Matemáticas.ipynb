Por lo tanto esto permite ver que la enseñanza de las matemáticas no siempre tiene una apropiación uniforme por parte del curso en general, y además este efecto se va acumulando año tras año, donde se traduce en mayor esfuerzo para los docentes y para los estudiantes de últimos cursos, el hecho de desarrollar las actividades satisfactoriamente.

```python
rendimiento_curso = df_general_ordenado.groupby('Curso')['% Total'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))

# 2. Diagrama de Barras del Promedio de % Total por Curso
# Usamos la columna 'Curso' como eje X y 'hue' para añadir un código de color a cada barra,
# aunque en este caso 'hue' solo sirve para asignar un color distinto a cada barra.
sns.barplot(
    data=rendimiento_curso,
    x='Curso',
    y='% Total',
    hue='Curso',  # Usamos 'Curso' como hue para diferenciar colores
    palette='viridis',
    legend=False  # Ocultamos la leyenda si es redundante
)

plt.title('Rendimiento Promedio (% Total) Clasificado por Curso')
plt.xlabel('Curso')
plt.ylabel('Promedio de % Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Rendimiento por curso de acuerdo a cada una de las habilidades
Es de tener en cuenta que el examen aplicado fue estandarizado para los cursos de sexto a undécimo, con temas sencillos, donde se pretendía que los estudiantes a través de la lectura y de la aplicación de los conocimientos analíticos, espaciales y lógicos pudiera obtener los mejores resultados, es muy interesante como el curso noveno se desempeño mucho mejor en el pensamiento analítico, en cambio los estudiantes de undécimo se desempeñaron mucho mejor espacialmente y lógicamente.

```python
# Crear un nuevo DataFrame solo con las columnas de interés
df_categorias = df_general_ordenado[['Curso', '% Analítica', '% Espacial', '% Lógica']]

# Transformar el DataFrame a formato "largo"
df_melted = df_categorias.melt(
    id_vars=['Curso'],
    value_vars=['% Analítica', '% Espacial', '% Lógica'],
    var_name='Habilidad',  # Nueva columna para el tipo de habilidad
    value_name='Porcentaje' # Nueva columna para el valor del porcentaje
)

# Agrupar para obtener el promedio de cada Habilidad DENTRO de cada Curso
df_promedios_habilidad = df_melted.groupby(['Curso', 'Habilidad'])['Porcentaje'].mean().reset_index()

plt.figure(figsize=(12, 7))

# Diagrama de Barras para comparar las Habilidades, separadas por 'Curso'
sns.barplot(
    data=df_promedios_habilidad,
    x='Habilidad',
    y='Porcentaje',
    hue='Curso',  # Usa 'Curso' como hue para generar barras agrupadas
    palette='Spectral'
)

plt.title('Rendimiento Promedio en Habilidades por Curso')
plt.xlabel('Habilidad')
plt.ylabel('Promedio de Porcentaje')
plt.legend(title='Curso', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
